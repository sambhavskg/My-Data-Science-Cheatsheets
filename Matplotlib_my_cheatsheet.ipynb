# Matplotlib 

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(1,11)
x
```

```python
y=3*x
y
```

```python
plt.plot(x,y)
plt.title('Line plot chart')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show
```

# changing line aesthetics

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=4)  # here colon is for dotted line
```

# adding two lines in the same plot

```python
x = np.arange(1,10)
y1 = 2*x
y2 = 4*x
print('x=',x)
print('y1=',y1)
print('y2=',y2)
```

```python
plt.plot(x,y1,color='y',linewidth=5,linestyle=':')
plt.plot(x,y2)
plt.grid(True)
plt.show()
```

# Adding line sub-plots

```python
x = np.arange(1,10)
y1 = 2*x
y2 = 4*x
print('x=',x)
print('y1=',y1)
print('y2=',y2)
```

```python
plt.subplot(1,2,1)   #means (one row needed,two coloum needed, index of subplot) 
plt.plot(x,y1,color='r',linewidth=5,linestyle=':')
plt.show()

plt.subplot(1,2,2)
plt.plot(x,y2,color='g', linewidth=5)
plt.show()
```

# Bar plot  ---- needed for categorial value distribution

```python
student = {'sambhav':30,'ananjay':50,'mayank':25,'payal':90}
```

```python
student.keys()
```

```python
student.values()
```

```python
st_name = list(student.keys())
st_marks = list(student.values())

plt.bar(st_name,st_marks)
plt.title('Student bar plot')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

# Box PLOT

### A Box and Whisker Plot (or Box Plot) is a convenient way of visually displaying the data distribution through their quartiles. It is a graphical rendition of statistical data based on the minimum, first quartile, median, third quartile, and maximum. The term “box plot” comes from the fact that the graph looks like a rectangle with lines extending from the top and bottom. Because of the extending lines, this type of graph is sometimes called a box-and-whisker plot.



# lower quartile = 25% value
# median = 50% 
# upper quartile = 75% value



```python
lst1 = [10,5,7.5,2.5]
plt.boxplot(lst1)
```

# pie charts -- we see the distribution between different categorial values

```python
names = ['apple','mango','lichi','banana','pineapple']
value = [20,50,69,86,45]
plt.pie(value,labels=names)
plt.show()
```

# changing aesthetics of pie chart

```python
names = ['apple','mango','lichi','banana','pineapple']
value = [20,50,69,86,45]
plt.pie(value,labels=names,autopct='%0.1f%%',colors=['blue','black','green','yellow','red'])
plt.show()
```

# Doughnut chart

```python
plt.pie(value,labels=names,radius=2)
plt.pie([100],radius=1,colors=['w'])
plt.show()
```
